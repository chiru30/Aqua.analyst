```python
plt.figure(figsize=(40,10))


sns.barplot(x=prod['State_Name'], y=prod['Production'])
```

```python
sns.heatmap(prod.corr(), annot = True)
```

```python
x=prod.groupby(['State_Name'])['Production','Prod/Area'].sum()
x.plot.pie(autopct='%1.lf%%', label=x.index,subplots=True,figsize=(30,30),startangle=90,textprops={'fontsize': 20})
plt.show()
```

```python
Top_10_production = prod.groupby("State_Name").Production.sum().nlargest(n =10)
```

```python
Top_10_production
```

```python
plt.style.use('seaborn')
Top_10_production.plot(kind ='bar', figsize =(10,10), fontsize =14,color= 'green')

plt.show()
```
