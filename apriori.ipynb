```python
def sublist(l1, l2):
    return set(l1) <= set(l2)
    
def check_subset_frequency(itemset, l, n):
    if n>1:    
        subsets = list(itertools.combinations(itemset, n))
    else:
        subsets = itemset
    for i in subsets:
        if not i in l:
            return False
    return True
```

```python
min_sup = 30
records = []

for i in range(len(bod_dis)):
    records.append([str(df_dis.values[i, j]) for j in range(7)])
    
items = sorted([item for sublist in records for item in sublist if item != 'nan'])
```

```python
#1st pass
```

```python
def pass_1(items, min_sup):
    c1 = {i:items.count(i) for i in items}
    l1 = {}
    
    for k, v in c1.items():
        if v >= min_sup:
            l1[k] = v
            
    return c1, l1
```

```python
c1, l1 = pass_1(items, min_sup)
l1
```

```python
#2nd pass
```

```python
def pass_2(l1, records, min_sup):
    l1 = sorted(list(l1.keys()))
    L1 = list(itertools.combinations(l1, 2))
    c2 = {}
    l2 = {}
    
    for i in L1:
        count = 0
        for j in records:
            if sublist(i, j):
                count += 1                
        c2[i] = count
    for k, v in c2.items():
        if v >= min_sup:
            if check_subset_frequency(k, l1, 1):
                l2[k] = v
    
    return c2, l2
```

```python
c2, l2 = pass_2(l1, records, min_sup)
l2
```

```python
#3rd pass¶
```

```python
def pass_3(l2, records, min_sup):
    l2 = sorted(list(l2.keys()))
    L2 = sorted(list(set([item for t in l2 for item in t])))
    L2 = list(itertools.combinations(L2, 3))
    c3 = {}
    l3 = {}
    
    for i in L2:
        count = 0
        for j in records:
            if sublist(i, j):
                count += 1                
        c3[i] = count
    for k, v in c3.items():
        if v >= min_sup:
            if check_subset_frequency(k, l2, 2):
                l3[k] = v
    
    return c3, l3
```

```python
c3, l3 = pass_3(l2, records, min_sup)
l3
```

```python
#4th pass
```

```python
def pass_4(l3, records, min_sup):
    l3 = sorted(list(l3.keys()))
    L3 = sorted(list(set([item for t in l3 for item in t])))
    L3 = list(itertools.combinations(L3, 4))
    c4 = {}
    l4 = {}
    
    for i in L3:
        count = 0
        for j in records:
            if sublist(i, j):
                count += 1                
        c4[i] = count
    for k, v in c4.items():
        if v >= min_sup:
            if check_subset_frequency(k, l3, 3):
                l4[k] = v
    
    return c4, l4
```

```python
c4, l4 = pass_4(l3, records, min_sup)
l4
```

```python
#5th pass
```

```python
def pass_5(l4, records, min_sup):
    l4 = sorted(list(l4.keys()))
    L4 = sorted(list(set([item for t in l4 for item in t])))
    L4 = list(itertools.combinations(L4, 5))
    c5 = {}
    l5 = {}
    
    for i in L4:
        count = 0
        for j in records:
            if sublist(i, j):
                count += 1                
        c5[i] = count
    for k, v in c5.items():
        if v >= min_sup:
            if check_subset_frequency(k, l4, 4):
                l5[k] = v
    
    return c5, l5
```

```python
c5, l5 = pass_5(l4, records, min_sup)
l5
```

```python
sets = []
for i in list(l3.keys()):
    subsets = list(itertools.combinations(i, 2))
    sets.append(subsets)
sets
```

```python
def support_count(itemset, itemlist):
    return itemlist[itemset]
```

```python
list_l3 = list(l3.keys())

for i in range(len(list_l3)):
    for j in sets[i]:
        a = j
        b = set(list_l3[i]) - set(j)
        confidence = (support_count(list_l3[i], itemlist)/support_count(j, itemlist))*100
        if confidence >= 95:
            print("{}->{} Confidence = ".format(a,b), confidence)
```
